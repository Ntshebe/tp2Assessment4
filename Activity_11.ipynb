```python
import pandas as pd 
from prophet import Prophet
from prophet.plot import plot_plotly
```

```python
df = pd.read_csv('spaza_data.csv')

df.columns=['ds','y']

model = Prophet()
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=365) # Forecasting 365 days
forecast = model.predict(future)
model.plot(forecast)
```
